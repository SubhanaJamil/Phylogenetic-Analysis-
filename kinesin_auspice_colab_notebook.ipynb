````python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import Phylo
import os
````

## 📥 Step 1: Sequence Data

For demonstration, we use **sample kinesin-like protein sequences**.  
In a real project, you can fetch sequences from NCBI using Biopython Entrez.


````python
# Create sample kinesin-related protein sequences (toy dataset)

sequences = [
 SeqRecord(Seq("MSTNPKPQRKTKRNTNRRPQDVKFPGGGK"), id="Kinesin_Human"),
 SeqRecord(Seq("MSTNPKPQRKTKRNTRRPPQDVKFPGA"), id="Kinesin_Mouse"),
 SeqRecord(Seq("MSSNPKPQRKAKRNTNRRPQDVKFPGGGK"), id="Kinesin14_Arabidopsis"),
 SeqRecord(Seq("MSTNPKPQRKTKRNTNRRPQEVKFPGGGK"), id="Kinesin14_yeast")
]

SeqIO.write(sequences, "kinesin.fasta", "fasta")
print("FASTA file created: kinesin.fasta")
````

## 🧪 Step 2: Multiple Sequence Alignment (MAFFT)

````python
!mafft --auto kinesin.fasta > aligned.fasta
print("Alignment complete: aligned.fasta")
````

## 🌳 Step 3: Phylogenetic Tree Construction (FastTree)

````python
!fasttree -protein aligned.fasta > kinesin_tree.nwk
print("Tree generated: kinesin_tree.nwk")
````

## 📊 Step 4: Preparing for Auspice

Next steps:

- Add metadata (species, protein family, etc.)



## 🔬 Interpretation

- Kinesin proteins cluster based on evolutionary similarity  
- Kinesin-14 forms a distinct subgroup
